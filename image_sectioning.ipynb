```python
import cv2
import numpy as np
from  skimage.metrics import structural_similarity

cap = cv2.VideoCapture('../VIPR Videos/Test03Cam02Trim.avi')
if (cap.isOpened() == False):
    print('Error opening file')

cap.set(cv2.CAP_PROP_POS_FRAMES, 450)    

ret, frame = cap.read()
# (B, G, R) = cv2.split(frame.astype('float'))
# rWindow = [R]
# gWindow = [G]
# bWindow = [B]

# total = 1

# window_size = 20
# quit = False

# i = 1
# while (cap.isOpened() and i < window_size and quit == False):
#     ret, frame = cap.read()
#     if ret:
#         (B, G, R) = cv2.split(frame.astype('float'))
#         rWindow.append(R)
#         gWindow.append(G)
#         bWindow.append(B)
#         i += 1
#         total += 1
        
#         cv2.imshow('frame', frame)
        
#         if cv2.waitKey(10) and 0xFF == ord('q'):
#             quit = True
#             break
#     else:
#         quit = True
#         break

# rAvg = sum(rWindow) / total
# gAvg = sum(gWindow) / total
# bAvg = sum(bWindow) / total

while (cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        cv2.imshow('frame', frame)
#         (B, G, R) = cv2.split(frame.astype('float'))
        
#         rAvg = (total * rAvg - rWindow[0]) / (total - 1)
#         gAvg = (total * gAvg - gWindow[0]) / (total - 1)
#         bAvg = (total * bAvg - bWindow[0]) / (total - 1)
#         total -= 1
        
#         rAvg = rAvg + (rWindow[total] - rAvg) / (total + 1)
#         gAvg = gAvg + (gWindow[total] - gAvg) / (total + 1)
#         bAvg = bAvg + (bWindow[total] - bAvg) / (total + 1)
#         total += 1
        
#         rWindow.append(R)
#         gWindow.append(G)
#         bWindow.append(B)
#         rWindow.pop(0)
#         gWindow.pop(0)
#         bWindow.pop(0)
        
#         avg = cv2.merge([bAvg, gAvg, rAvg]).astype('uint8')
#         cv2.imshow('average', avg)
        
        num_sections = 6
        section_width = int(frame.shape[1] / num_sections)
        norms = frame.copy()
        threshes = frame.copy()
        for i in range(num_sections):
            img = frame[:, i * section_width : (i+1) * section_width]
            img = cv2.GaussianBlur(img, (3, 3), 0)
#             cv2.imshow('{} section'.format(i+1), img)
            
            max_val = np.amax(img)
            img = img.astype('float32') / max_val * 255.0
            img = img.astype('uint8')
#             cv2.imshow('{} norm'.format(i+1), img)
            norms[:, i * section_width : (i+1) * section_width] = img
            
            mean = np.mean(img)
            std = np.std(img)
            
            _, img = cv2.threshold(img, mean + 3*std, 255, cv2.THRESH_TOZERO)
#             cv2.imshow('{} thresh'.format(i+1), img)
            threshes[:, i * section_width : (i+1) * section_width] = img
        
        cv2.imshow('norms', norms)
        cv2.imshow('threshes', threshes)
            
        key = cv2.waitKey(30)
        if key & 0xFF == ord('r'):
            cap.set(cv2.CAP_PROP_POS_FRAMES, 450)
        elif key & 0xFF == ord('q'):
            quit = True
            break
    else:
        quit = True
        break
        
cap.release()
cv2.destroyAllWindows()
```
