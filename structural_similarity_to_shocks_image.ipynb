```python
import cv2
import numpy as np
from  skimage.metrics import structural_similarity

shocks = cv2.imread('rocket_test.jpg')
shocks = cv2.cvtColor(shocks, cv2.COLOR_BGR2GRAY)
cap = cv2.VideoCapture('../VIPR Videos/Test03Cam02Trim.avi')
if (cap.isOpened() == False):
    print('Error opening file')

ret, frame = cap.read()
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

height = frame.shape[0]
width = frame.shape[1]
shocks = cv2.resize(shocks, (width, height))
cv2.imshow('shocks image', shocks)
max_sim = -1
max_sim_frame = None

while (cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        similarity = structural_similarity(shocks, frame, multichannel=True)
        if similarity > max_sim:
            max_sim = similarity
            max_sim_frame = frame
            cv2.imshow('max similarity', max_sim_frame)
            print('max similarity: {}'.format(similarity))
            
        cv2.imshow('frame similarity', frame)
        print('frame similarity: {}'.format(similarity))
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    else:
        break
        
cap.release()
cv2.destroyAllWindows()
```
