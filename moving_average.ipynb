```python
import cv2
import numpy as np
from  skimage.metrics import structural_similarity

cap = cv2.VideoCapture('../VIPR Videos/Test03Cam02Trim.avi')
if (cap.isOpened() == False):
    print('Error opening file')

cap.set(cv2.CAP_PROP_POS_FRAMES, 550)    

ret, frame = cap.read()
(B, G, R) = cv2.split(frame.astype('float'))
rWindow = [R]
gWindow = [G]
bWindow = [B]

total = 1

window_size = 20
quit = False

i = 1
while (cap.isOpened() and i < window_size and quit == False):
    ret, frame = cap.read()
    if ret:
        (B, G, R) = cv2.split(frame.astype('float'))
        rWindow.append(R)
        gWindow.append(G)
        bWindow.append(B)
        i += 1
        total += 1
        
        cv2.imshow('frame', frame)
        
        if cv2.waitKey(10) and 0xFF == ord('q'):
            quit = True
            break
    else:
        quit = True
        break

rAvg = sum(rWindow) / total
gAvg = sum(gWindow) / total
bAvg = sum(bWindow) / total

th1 = 225
th2 = 170

while (cap.isOpened() and quit == False):
    ret, frame = cap.read()
    if ret:
        cv2.imshow('frame', frame)
        (B, G, R) = cv2.split(frame.astype('float'))
        
        rAvg = (total * rAvg - rWindow[0]) / (total - 1)
        gAvg = (total * gAvg - gWindow[0]) / (total - 1)
        bAvg = (total * bAvg - bWindow[0]) / (total - 1)
        total -= 1
        
        rAvg = rAvg + (rWindow[total] - rAvg) / (total + 1)
        gAvg = gAvg + (gWindow[total] - gAvg) / (total + 1)
        bAvg = bAvg + (bWindow[total] - bAvg) / (total + 1)
        total += 1
        
        rWindow.append(R)
        gWindow.append(G)
        bWindow.append(B)
        rWindow.pop(0)
        gWindow.pop(0)
        bWindow.pop(0)
        
        avg = cv2.merge([bAvg, gAvg, rAvg]).astype('uint8')
        cv2.imshow('average', avg)
        
        _, thresh1 = cv2.threshold(avg, th1, 255, cv2.THRESH_TOZERO_INV)
        _, thresh2 = cv2.threshold(thresh1, th2, 255, cv2.THRESH_TOZERO)
        cv2.imshow('threshold', thresh2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            quit = True
            break
    else:
        quit = True
        break
        
cap.release()
cv2.destroyAllWindows()
```
